```python
import pandas as pd

names1880 = pd.read_csv('babynames/yob1880.txt', names=['names', 'sex', 'births'])
names1880
```

```python
names1880.groupby('sex').births.sum()
```

```python
years = range(1880, 2011)
pieces = []
columns = ['names', 'sex', 'births']
for year in years:
    path = 'babynames/yob%d.txt'%year
    frame = pd.read_csv(path, names=columns)
    
    frame['year'] = year
    pieces.append(frame)

names = pd.concat(pieces, ignore_index=True)
```

```python
names
```

```python
total_birth = names.pivot_table('births', index='year', columns='sex',aggfunc=sum)
total_birth.tail
```

```python
total_birth.plot(title='Total births by sex and year')
```

```python
def add_prob(group):
    group['prop'] = group.births / group.births.sum()
    return group

names = names.groupby(['year', 'sex']).apply(add_prob)
names
```

```python
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]

grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
top1000
```

```python
top1000.reset_index(inplace=True, drop=True)
top1000
```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
```

```python
total_births = top1000.pivot_table('births', index='year', columns='names', aggfunc=sum)
total_births.info()
```

```python
subset = total_births[['John', 'Harry', 'Mary', 'Marilyn']]
subset.plot(subplots=True, figsize=(12, 10), grid=False, title="Number of birth per year")
```

```python
import numpy as np
table = top1000.pivot_table('prop', index='year', columns='sex', aggfunc=sum)
table.plot(title='Sum of table1000.prop by year and sex', yticks=np.linspace(0, 1.2,  13), xticks=range(1880, 2020, 10))
```

```python
df = boys[boys.year ==  2010 ]
df
```

```python
prop_cumsum = df.sort_values(by='prop', ascending=False).prop.cumsum()
prop_cumsum[:10]
```

```python
prop_cumsum.values.searchsorted(0.5)
```

```python
df = boys[boys.year == 1900]
in1900 = df.sort_values(by='prop', ascending=False).prop.cumsum()
in1900.values.searchsorted(0.5) + 1
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_values(by='prop',  ascending=False)
    return group.prop.cumsum().values.searchsorted(q) + 1
diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
```

```python
diversity.head()
```

```python
diversity.plot(title="Number og popular names in top 50%")
```

```python
get_last_letter = lambda x : x[-1]
last_letters = names.names.map(get_last_letter)
last_letters.name = 'last_letter'

table = names.pivot_table('births', index=last_letters, columns=['sex', 'year'], aggfunc=sum)
```

```python
subtable = table.reindex(columns=[1910, 1960, 2010], level='year')
subtable.head()
```

```python
subtable.sum()
```

```python
letter_prop = subtable/ subtable.sum()
letter_prop
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(10, 8))
letter_prop['M'].plot(kind='bar', rot=0, ax=axes[0], title="Male")
letter_prop['F'].plot(kind='bar', rot=0, ax=axes[1], title="FeMale", legend=False)

```

```python
letter_prop = table / table.sum()
dny_ts = letter_prop.loc[['d', 'n', 'y'], 'M'].T
dny_ts
```

```python
dny_ts.plot()
```

```python
all_names = pd.Series(top1000.names.unique())
```

```python
lesley_like = all_names[all_names.str.lower().str.contains('lesl')]
lesley_like
```

```python
filtered = top1000[top1000.names.isin(lesley_like)]
filtered.groupby('names').births.sum()
```

```python
table = filtered.pivot_table('births',index='year', columns='sex',  aggfunc='sum')
table = table.div(table.sum(1), axis=0)
table.tail()
```

```python
table.plot(style={'M':'k-', 'F':'k--'})
```
