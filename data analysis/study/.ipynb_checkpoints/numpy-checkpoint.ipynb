```python
points = np.arange(-5, 5, 0.01) 
print(points)
```

```python
ys
```

```python
z = np.sqrt(xs**2 + ys**2)
z
```

```python
import random
position=0
walk = [position]
steps=1000
for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)
```

```python
walk
```

```python
plt.plot(walk[:100])
```
